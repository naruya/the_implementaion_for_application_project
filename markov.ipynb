```python
import numpy as np
import matplotlib.pyplot as plt
import math
from numpy.random import *
```

```python
N = 1000
count = 0
random = rand(N, 2)
plt.figure(figsize=(5, 5))
for i in range(N):
    tmp = random[i, 0]**2 + random[i, 1]
    if tmp <= 1:
        count += 1
        plt.scatter(random[i, 0], random[i, 1], color="r")
    else:
        plt.scatter(random[i, 0], random[i, 1], color="b")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.plot()

def return_count(N):
    count = 0
    random = rand(N, 2)
    for i in range(N):
        tmp = random[i, 0]**2 + random[i, 1]
        if tmp**0.5 <= 1:
            count += 1
    return count / N
```

```python
print(count / N)
```

```python
ans = []
plt.figure(figsize=(3, 3))
N_list = [10 ** (i + 1) for i in range(7)]
for n in N_list:
    result_pi = math.pi
    ans_ = math.log(abs(result_pi - return_count(n)))
    ans.append(ans_)
plt.scatter(np.log(N_list), ans)
plt.show()
```
